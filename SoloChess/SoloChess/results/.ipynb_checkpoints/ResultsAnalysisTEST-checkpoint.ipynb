### CSV Merging

```python
## Load .csv files
char = pd.read_csv('characteristicsTEST.csv')
rand = pd.read_csv('randomTEST.csv')
rank = pd.read_csv('rankTEST.csv')
attack = pd.read_csv('attackTEST.csv')
center = pd.read_csv('centerTEST.csv')

## Create full dataframe
all_data = ft.reduce(lambda  left,right: pd.merge(left,right,on=['id'], how='outer'), [rand, rank, attack, center, char])
all_data.to_csv('alldataTEST.csv', index=False)
```

### Difficulty N

```python
df = pd.read_csv('alldataTEST.csv')

## Disregard columns
df = df.drop(["time_random", "time_rank", "time_attack", "time_center",
             'total_rank', 'sd_rank', 'total_attack', 'sd_attack','total_center', 'sd_center'], axis=1)

## Group by N
df = df.groupby(['N'], as_index=False).mean().drop(['id'], axis=1)

## Add columns with normalized number of backtracks compared to bt_random
def normalize(df, row, by):
    df[row+"_norm"] = df[row] / df[by]
    
for row in ['bt_random','bt_rank', 'bt_attack', 'bt_center']:
    normalize(df, row, 'bt_random')

df
```

```python
## Plots N vs. total number of backtracks
df.plot(x="N", y=["bt_random", "bt_rank", "bt_attack", "bt_center"], kind="line")
df.plot(x="N", y=["bt_random", "bt_rank", "bt_attack", "bt_center"], kind="bar")
df.plot(x="N", y=["bt_random", "bt_rank", "bt_attack", "bt_center"], kind="box")
```

```python
## Plot N vs. normalized number of backtracks
df.plot(x="N", y=["bt_random_norm", "bt_rank_norm", "bt_attack_norm", "bt_center_norm"], kind="line")
df.plot(x="N", y=["bt_random_norm", "bt_rank_norm", "bt_attack_norm", "bt_center_norm"], kind="bar")
df.plot(x="N", y=["bt_random_norm", "bt_rank_norm", "bt_attack_norm", "bt_center_norm"], kind="box")
```
