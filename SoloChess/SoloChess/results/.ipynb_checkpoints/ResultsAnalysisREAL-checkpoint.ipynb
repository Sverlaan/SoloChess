# Solo Chess - Data Analysis

```python
## Imports
import numpy as np
import pandas as pd
import statsmodels
import statistics
import matplotlib.pyplot as plt
import re
import functools as ft
from scipy import stats
```

### Loading Data

```python
## Load .csv files
char = pd.read_csv('characteristics.csv')
rand = pd.read_csv('random2-14.csv')
rank = pd.read_csv('rank2-14.csv')
attack = pd.read_csv('attack2-14NEWNEW.csv')
center = pd.read_csv('center2-14.csv')

## Create full dataframe
df = ft.reduce(lambda  left,right: pd.merge(left,right,on=['id'], how='outer'), [rand, rank, attack, center, char])

## Disregard columns
df = df.drop(["time_random", "time_rank", "time_attack", "time_center"], axis=1)

## Remove puzzles with N < 4
df = df[df['N'] > 3]

df.to_csv('alldata.csv', index=False)
df
```

### Compare on number of pieces $N$

```python
## Load data
data = pd.read_csv('alldata.csv')

## Disregard columns
data = data.drop(["rankdist"], axis=1)

## Group by N
data_by_N = data.groupby(['N'], as_index=False).mean().drop(['id'], axis=1)

## Add columns with normalized number of backtracks compared to bt_random
def normalize(df, col, by):
    df[col+"_norm"] = df[col] / df[by]
    
for col in ['bt_random', 'bt_rank', 'bt_attack', 'bt_center']:
    normalize(data_by_N, col, 'bt_random')
    
data_by_N.set_index('N')
```

```python
## Plots N vs. total number of backtracks
data_by_N.plot(x="N", y=["bt_random", "bt_rank", "bt_attack", "bt_center"], kind="line")
plt.xticks(data_by_N["N"])
plt.show()
```

```python
data_by_N.plot(x="N", y=["bt_random_norm", "bt_rank_norm", "bt_attack_norm", "bt_center_norm"], 
               label=["random", "rank", "attack", "center"],
               kind="line", figsize=(6,4), grid=True)

plt.xticks(data_by_N["N"])
plt.title("Effect heuristics")
plt.xlabel("N")
plt.ylabel("backtrack ratio")
plt.savefig('figures\\lines.png')
plt.show()
```

```python
data_by_N.plot(x="N", y=["bt_random_norm", "bt_rank_norm", "bt_attack_norm", "bt_center_norm"],
               label=["random", "rank", "attack", "center"], kind="bar", figsize=(8,4))

plt.xticks(rotation=45)
#plt.title("Mince Pie Consumption Totals")
plt.xlabel("N")
plt.ylabel("backtrack ratio")
plt.savefig('figures\\bar.png')
plt.show()
```

```python
data_by_N.boxplot(column=['bt_rank_norm', 'bt_attack_norm', 'bt_center_norm'],
                  labels=['a', 'b', 'c'], figsize=(4,5))

#plt.title("Mince Pie Consumption Totals")
plt.ylabel("backtrack ratio")
plt.savefig('figures\\box.png')
plt.show()
```

### Characteristics

```python
data = pd.read_csv('alldata.csv')
    
arr = np.zeros((len(data['rankdist'])))
for i, ranking in enumerate(data['rankdist']):
    skewie = stats.skew([float(a) for a in str(ranking)])
    arr[i] = skewie
data['skew'] = arr
```

```python
data['skew_bin'] = pd.cut(data['skew'],[-3, -0.5, 0.5, 3], include_lowest=True)

data.groupby(['skew_bin']).size().reset_index(name='counts')
```

```python
arr = np.zeros((len(data['rankdist'])))
for i, ranking in enumerate(data['rankdist']):
    rankings = [int(a) for a in str(ranking)]
    arr[i] = sum(rankings) / len(rankings)
data['mean_rank'] = arr

data["mean_rank"].sum() / len(data["mean_rank"].values)
```

```python
data = data.groupby(['skew_bin'], as_index=False).mean()
data

## Add columns with normalized number of backtracks compared to bt_random
def normalize(df, col, by):
    df[col] = df[col] / df[by]

for col in ['bt_rank', 'bt_attack', 'bt_center', 'bt_random']:
    normalize(data, col, 'bt_random')

data.plot(x='skew_bin', y=["bt_random", "bt_rank", "bt_attack", "bt_center"], kind="line", figsize=(6,4))
```
